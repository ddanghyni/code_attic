### 3.2 Object-Oriented Design for Implementation

```python
import time
import numpy as np
import torch
from torch import nn
from d2l import torch as d2l
```

```python
'''이미 선언된 class에 method를 추가하는 방법'''

def add_to_class(Class):
    """Register functions as method in created class"""
    def wrapper(obj):
        setattr(Class, obj.__name__, obj)
    return wrapper
```

```python
'''example'''
class A:
    def __init__(self):
        self.b = 1

a = A()
```

```python
'''이제 class A에 do라는 method 추가'''

@add_to_class(A)
def do(self):
    print(f"Class attribute {self.b}")

a.do()
```
